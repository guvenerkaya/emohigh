```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import pickle
%matplotlib inline
```

```python
arousal_X = pd.read_csv('/Volumes/LIRIS_ACCEDE/LIRIS-ACCEDE-features/ACCEDEfeaturesArousal_TAC2015.txt',delimiter="\t", header=0, index_col=[0,1])
pd.options.display.max_rows = 9799
pd.options.display.max_columns = 30
y = pd.read_csv('/Volumes/LIRIS_ACCEDE/LIRIS-ACCEDE-annotations/annotations/ACCEDEranking.txt' ,delimiter="\t", header=0, index_col=[0,1], usecols=[0,1,4,5])
arousal_y = y.iloc[:,1]
df_1 = pd.concat([arousal_X,arousal_y], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(arousal_X, arousal_y, test_size=0.5, random_state=0)
```

```python
X_train_1.shape
#X_train_1
```

```python
lr_1 = LinearRegression()
lr_1.fit(X_train_1, y_train_1)
```

```python
y_pred_1 = lr_1.predict(X_test_1)
#y_train_1
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test_1,y_pred_1)
print(score)
from sklearn.metrics import mean_squared_error
import scipy
scipy.stats.pearsonr(y_test_1, y_pred_1)
#mes = mean_squared_error(y_test_1,y_pred_1)
#print(mes)
result = pd.DataFrame({'Actual': y_test_1, 'Predicted': y_pred_1})
result
```
